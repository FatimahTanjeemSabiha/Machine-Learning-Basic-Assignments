```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

fts = pd.read_csv('/content/drive/MyDrive/data.csv')
fts.head()

x = fts.iloc[:,1:3]
x.head()

plt.scatter(fts['Model'],fts['Engine Fuel Type'])

kmeans = KMeans(3)
kmeans.fit(x)

identified_clusters = kmeans.fit_predict(x)
identified_clusters


data_with_clusters = fts.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['Model'],data_with_clusters['Engine Fuel Type'])

wcss=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

plt.plot(range(1,7),wcss)
```
